```python
import pandas as pd
import numpy as np
import chart_studio.plotly as py
import seaborn as sns
import cufflinks as cf
import plotly.express as px
%matplotlib inline

from plotly.offline import *
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
arr_1 = np.random.randn(50,4)
df_1 = pd.DataFrame(arr_1, columns=['A','B','C','D'])
df = pd.read_csv('dataIntel.csv')
df_1.head()
```

```python
df_1.iplot()
```

```python
import plotly.graph_objects as go
df_stocks = px.data.stocks()
px.line(df_stocks, x='date',y=['GOOG','AAPL','MSFT'],labels={'x':'Date','y':'Stock Prices'},title='Google vs Apple vs Microsoft')
```

```python
px.line(df,x='NODEID',y=['TEMPERATURE','HUMIDITY'],labels={'x':'NODEID','y':'Features'},title='TEMPERATURE vs HUMIDITY')
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_stocks.date,y=df_stocks.AAPL,name='Apple',mode='lines'))
fig.add_trace(go.Scatter(x=df_stocks.date,y=df_stocks.AMZN,name='Amazon',mode='lines+markers'))
fig.add_trace(go.Scatter(x=df_stocks.date,y=df_stocks.GOOG,name='Google',mode='lines+markers',
line=dict(color='firebrick',width=2,dash='dashdot')))
fig.update_layout(title='Apple vs Amazon vs Google',xaxis_rangeslider_visible=True)
```
