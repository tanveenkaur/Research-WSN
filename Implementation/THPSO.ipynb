```python
class Particle:
    def __init__(self, x, y, vx, vy, fitness):
        self.x = x
        self.y = y
        self.vx = vx
        self.vy = vy
        self.fitness = fitness
        self.pbest_x = x
        self.pbest_y = y
        self.pbest_fitness = fitness
        self.gbest_x = x
        self.gbest_y = y
        self.gbest_fitness = fitness
    
```

```python
def PSO(particles, max_iterations, w, c1, c2):
    for i in range(max_iterations):
        for j in range(len(particles)):
            # update velocity
            particles[j].vx = w * particles[j].vx + c1 * random.random() * (particles[j].pbest_x - particles[j].x) + c2 * random.random() * (particles[j].gbest_x - particles[j].x)
            particles[j].vy = w * particles[j].vy + c1 * random.random() * (particles[j].pbest_y - particles[j].y) + c2 * random.random() * (particles[j].gbest_y - particles[j].y)
            # update position
            particles[j].x = particles[j].x + particles[j].vx
            particles[j].y = particles[j].y + particles[j].vy
            # update fitness
            particles[j].fitness = fitness(particles[j].x, particles[j].y)
            # update pbest
            if particles[j].fitness < particles[j].pbest_fitness:
                particles[j].pbest_x = particles[j].x
                particles[j].pbest_y = particles[j].y
                particles[j].pbest_fitness = particles[j].fitness

            # update gbest
            if particles[j].fitness < particles[j].gbest_fitness:
                particles[j].gbest_x = particles[j].x
                particles[j].gbest_y = particles[j].y
                particles[j].gbest_fitness = particles[j].fitness
    return particles
```

```python
A = 0.5

# base station coordinates
x_bs = 0
y_bs = 0

def CH2BS(x_ch,y_ch,x_bs,y_bs):
    return math.sqrt((x_ch - x_bs)**2 + (y_ch - y_bs)**2)

def CH2C(x_ch,y_ch,x_c,y_c):
    return math.sqrt((x_ch - x_c)**2 + (y_ch - y_c)**2)

def dMax(XCH,YCH,N):
    d_max = 0
    for i in range(N):
        d = CH2BS(XCH[i],YCH[i],x_bs,y_bs)
        if d > d_max:
            d_max = d
        
    return d_max + 1


def d2Max(XCH,YCH,x_c,y_c,N):
    d_max = 0
    for i in range(N):
        d = CH2C(XCH[i],YCH[i],x_c,y_c)
        if d > d_max:
            d_max = d
        
    return d_max + 1




```
